```python
annee_true = [int(i) for i in annee if i != '']
```

```python
print(min(annee_true))
print(max(annee_true))
```

```python
trees = 0
trees_per_year = []
for year in sorted(set(annee_true)):
    print(f'{year} : {annee_true.count(year)}')
    trees_per_year.append((str(year), annee_true.count(year)))
    trees += annee_true.count(year)
```

```python
print(f'mean : {round(trees/(max(annee_true)-min(annee_true)))} per year')
```

```python
trees_per_year
```

```python
maires_liste = ['Leon_Martin', 'Albert_Michallon', 'Hubert_Dubedout', 'Alain_Carignon', 'Michel_destot', 'Eric_Piolle']
mandats = [(1949, 1959), (1959, 1965), (1965, 1983), (1983, 1995), (1995, 2014), (2014, 2021)]

maires = {}

for name, mandat in zip(maires_liste, mandats):
    maires[name] = {'mandat':mandat, 'trees':0}
```

```python
maires['Leon_Martin']
```

```python
for year in trees_per_year:
    for maire in maires:
        if maires[maire]['mandat'][0] < int(year[0]) <= maires[maire]['mandat'][1]:
            maires[maire]['trees'] += year[1]
```
